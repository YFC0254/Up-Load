```python
import SignalProcess1
import numpy as np
# 测试SignalProcess1里面的AntiAliasingFilter类

# 信号数据
data = np.arange(0, 100, 0.1)
# 采样频率
sample_rate = 10
# 截止频率
cutoff_frequency = 1
# 滤波器阶数
filter_order = 4
# 低通滤波
filtered_data = SignalProcess1.AntiAliasingFilter().apply_lowpass_filter(data, cutoff_frequency, sample_rate, filter_order)
print(filtered_data)
# 画出滤波后的波形
SignalProcess1.BasicSignal(filtered_data, sample_rate).draw('filtered_data', 'time', 'amplitude')
# 画出滤波后的功率谱
SignalProcess1.BasicSignal(filtered_data, sample_rate).draw_psd('filtered_data')

```
